```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints=np.array([1,8])
ypoints=np.array([3,10])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints=np.array([1,8])
ypoints=np.array([3,10])

plt.plot(xpoints,ypoints,'o')
plt.show()
```

```python
xpoints=np.array([1,2,6,8])
ypoints=np.array([3,8,1,10])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
ypoints=np.array([3,8,1,10])

plt.plot(ypoints)
plt.show()
```

## Markers

```python
ypoints=np.array([3,8,1,10])

plt.plot(ypoints, marker='d')
plt.show()
```
