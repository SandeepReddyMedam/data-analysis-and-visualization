# Indexing and slicing

## Indexing

```python
a=np.arange(10)
print(a)
print(a[5])
print(a[-1])
```

```python
b=np.diag([1,2,3])
print(b)
print(b[2,2])
```

```python
b[2,1]=10#assining a value to index
print(b)
```

## Slicing

```python
a=np.arange(10)
print(a[1:10:2])
```

```python
b=np.arange(10)
b[5:]=10#assigns 10 from index 5 to end
print(b)
```

# Copies and Viwes

```python
a=np.arange(10)
b=a[::2]
np.shares_memory(a,b)
```

```python
b[0]=10
print(b)
print(a)#a is also updated, since it shares memory location
```

```python
c=a[::2].copy()#force the copy
np.shares_memory(a,c)
```

```python
c[0]=5
print(c)
print(a)
```

# Fancy Indexing

## Using Boolean Mask

```python
a=np.random.randint(0,20,15)
print(a)
```

```python
mask=(a % 2 == 0)
```

```python
even_numbers=a[mask]
print(even_numbers)
```

```python
a[mask]=-1
print(a)
```
